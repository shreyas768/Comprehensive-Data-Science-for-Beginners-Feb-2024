```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cancer_tumor_data_features.csv')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc_X = sc.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(sc_X)
```

```python
plt.scatter(principal_components[:,0],principal_components[:,1])
```

```python
pca.n_components_
```

```python
pca.components_
```

```python
df_pc = pd.DataFrame(pca.components_,index=['PC1','PC2'],columns = df.columns)
```

```python
plt.figure(figsize=(20,5))
sns.heatmap(df_pc)
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
pca30 = PCA(n_components=30)
pca30.fit(sc_X)
```

```python
np.sum(pca30.explained_variance_ratio_)
```

```python
explained_variance = []

for n in range(1,30):
    pca1 = PCA(n_components=n)
    pca1.fit(sc_X)
    explained_variance.append(np.sum(pca1.explained_variance_ratio_))
```

```python
explained_variance
```

```python
plt.plot(range(1,30),explained_variance)
plt.xlabel('Number of PC')
plt.ylabel('Variance Explained')
```
