```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,15)
```

```python
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
```

```python
plt.figure(figsize=(12,6),dpi=200)
plt.plot(x,y,color='m', marker='D', linestyle='-.',linewidth=2, markersize=5)
plt.xlabel('X axis title here')
plt.ylabel('Y axis title here')
plt.title('Figure title here')
plt.xlim(2,6)
plt.ylim(0,12)
#plt.show()
plt.savefig('example.png')
```
