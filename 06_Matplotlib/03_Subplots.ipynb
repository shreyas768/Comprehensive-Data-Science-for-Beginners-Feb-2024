```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2*x
```

### SubPlots 

```python
fig, axes = plt.subplots(nrows=1,ncols=2) #Empty canvas
axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2) #Empty canvas
axes[0,0].plot(a,b)
axes[0,0].set_title('0 0 Title')
axes[0,0].set_xlabel('0 0 xlabel')

axes[0,1].scatter(x,y)

plt.tight_layout()

fig.savefig('example.png')
```
