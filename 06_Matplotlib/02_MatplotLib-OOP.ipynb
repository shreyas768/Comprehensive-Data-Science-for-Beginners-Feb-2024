```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
y = 2*x
x
```

```python
y
```

## Object Oriented Programming (OOP) method

```python
fig = plt.figure()  #create figure (empty canvas)

#add set of axes to figure

ax = fig.add_axes([0,0,1,1])  # Left,Bottom,Width,Height (range 0 to 1)

#plot on that set of axes

ax.plot(x,y)

plt.xlim(0,9)
plt.ylim(0,18)
plt.show()
```

```python
fig = plt.figure()  #create figure (empty canvas)

#add set of axes to figure

ax = fig.add_axes([0,0,1,1])  # Left,Bottom,Width,Height (range 0 to 1)

#plot on that set of axes

ax.plot(a,b)

plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])  #Larger Figure
ax2 = fig.add_axes([0.2,0.5,0.3,0.3]) #Smaller Figure
ax3 = fig.add_axes([1,1,0.3,0.3])

ax1.plot(a,b, label = 'this is 4th power')
ax1.plot(x,y, label = 'y = 2*x')
ax1.set_xlabel('X Label')
ax1.set_ylabel('Y label')
ax1.set_xlim(0,8)
ax1.set_ylim(0,100)
ax1.set_title('Big Figure Zoomed In')
ax1.legend(loc = 2)  #loc 1 = Upper right corner, loc2= upper left corner
#loc3 = lower left corner #loc4=lower right corner
#loc0 = let matplotlib to decide the optimal location

ax2.plot(x,y)
ax2.set_xlabel('X Label')
ax2.set_ylabel('Y label')
ax2.set_title('Small Figure')


ax3.plot(a,b)
ax3.set_xlabel('X Label')
ax3.set_ylabel('Y label')
ax3.set_title('Small Figure 2')
```
