```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
#countplot
plt.figure(figsize=(10,4))
sns.countplot(x='division',data=df,hue='training level',palette='cubehelix')
```

```python
#barplot
plt.figure(figsize=(10,4))
sns.barplot(x='level of education',y='salary',data=df,estimator=np.median, hue = 'training level')
```
