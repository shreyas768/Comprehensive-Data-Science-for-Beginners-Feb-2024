```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
#Rugplot
sns.rugplot(x='salary',data=df,height=0.4, hue = 'division')
```

```python
#displot

sns.displot(data=df,x='salary')
```

```python
sns.histplot(data=df,x='salary',kde=True, bins=10, color='red', legend = True, ls = '--', edgecolor='blue')
```

```python
sns.kdeplot(data=df,x='salary')
```
