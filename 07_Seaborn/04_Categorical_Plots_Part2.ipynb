```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df['race/ethnicity'].value_counts()
```

```python
df.shape
```

```python
df.info()
```

```python
#Boxplot
plt.figure(figsize=(12,5))
sns.boxplot(x='parental level of education',y='reading score',data=df, hue = 'gender')
```

```python
#Boxplot
plt.figure(figsize=(12,20))
sns.boxplot(y='parental level of education',x='reading score',data=df, hue = 'race/ethnicity', orient='h')
```

```python
#Violinplot
plt.figure(figsize=(12,5))
sns.violinplot(x='parental level of education',y='reading score',data=df, hue = 'gender',split=True,
               inner = "quartile")
```

```python
#swarmplot
sns.swarmplot(x='writing score',data=df)
```

```python
sns.swarmplot(x='writing score',y = 'gender',data=df, hue = 'race/ethnicity', dodge = True)
```
