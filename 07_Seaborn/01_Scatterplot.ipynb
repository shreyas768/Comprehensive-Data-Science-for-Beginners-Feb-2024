```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
#hue : color points based on some categorical feature in the DataFrame

plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df, hue = 'division')
plt.show()
```

```python
plt.figure(figsize=(12,9))
sns.scatterplot(x='salary',y='sales',data=df, hue = 'division',size='training level', style='level of education')
plt.legend(title_fontsize='xx-small')
plt.show()
```
