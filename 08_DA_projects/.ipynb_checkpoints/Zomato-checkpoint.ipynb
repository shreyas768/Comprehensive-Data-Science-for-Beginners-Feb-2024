```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['address','phone', 'dish_liked','cuisines', 'reviews_list', 'menu_item','listed_in(city)','listed_in(city)'],axis=1)
df.head()
```

```python
df['book_table'].value_counts()
```

```python
df.info()
```

```python
df['rate'].unique()
```

```python
def ratings_cleaned(val):
    if (val=='NEW' or val == '-'):
        return np.nan
    else:
        return float(str(val).split('/')[0])
```

```python
df['rate'] = df['rate'].apply(ratings_cleaned)
```

```python
df.isnull().sum()
```

```python
df['rate'].fillna(np.round(df['rate'].mean(),1),inplace=True)
```

```python
df['rate'].value_counts()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
df['approx_cost(for two people)'].unique()
```

```python
def cleaned_approx_cost(num):
    num = str(num)
    if ',' in num:
        num =int(num.replace(',',''))
        return num
    else:
        return int(num)
    
df['approx_cost(for two people)']=df['approx_cost(for two people)'].apply(cleaned_approx_cost)
df['approx_cost(for two people)'].unique()
```

```python
rest_type_df = df['rest_type'].value_counts()
rest_type_df_below1000 = rest_type_df[rest_type_df<1000]
rest_type_df_below1000
```

```python
def cleaned_rest_type(tp):
    if (tp in rest_type_df_below1000):
        return 'Others'
    else:
        return tp
    
df['rest_type']=df['rest_type'].apply(cleaned_rest_type)
df['rest_type'].value_counts()
```

```python
res_loc = df['location'].value_counts()
res_loc_below250 = res_loc[res_loc<250]
res_loc_below250
```

```python
def cleaned_location(loc1):
    if loc1 in res_loc_below250:
        return 'Others'
    else:
        return loc1
```

```python
df['location']=df['location'].apply(cleaned_location)
df['location'].value_counts()
```

```python
df['listed_in(type)'].value_counts()
```

```python
df['name'].value_counts()[:10]
```

```python
rest_cost = df['approx_cost(for two people)'].groupby(df['name'],sort=True)
dct1 = {}

for i,j in df['name'].value_counts()[:10].to_dict().items():
    dct1[i] = np.round(rest_cost.get_group(i).mean(),2)
    
cost_df = pd.DataFrame(list(dct1.items()),columns = ['rest_name','avg_cost_two_ppl'])
cost_df

sns.barplot(data = cost_df.sort_values(by=['avg_cost_two_ppl'], ascending=False), 
            x='avg_cost_two_ppl',y='rest_name')
plt.title('Average Cost of two people - Top 10 restaurants of Bengaluru');
```
