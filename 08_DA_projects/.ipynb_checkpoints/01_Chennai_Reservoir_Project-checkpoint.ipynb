```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('chennai_reservoir_levels.csv')
df
```

```python
#plotly, cufflinks libraries
#tableau / power BI
#SQL
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='%d-%m-%Y')
```

```python
df.info()
```

```python
df['Year'] = df['Date'].apply(lambda x:x.year)
df['Month'] = df['Date'].apply(lambda x:x.month)
```

```python
df.head()
```

```python
df['Total'] = df[[ 'POONDI', 'CHOLAVARAM', 'REDHILLS', 'CHEMBARAMBAKKAM']].sum(axis=1)
df.head()
```

```python
df.groupby('Year', as_index=False).mean()
```

```python
Yearwise = df.groupby('Year', as_index=False).mean()
Yearwise = Yearwise[['Year', 'POONDI', 'CHOLAVARAM', 'REDHILLS', 'CHEMBARAMBAKKAM','Total']]
Yearwise
```

```python
plt.figure(figsize=(14,8))
sns.barplot(x='Year',y='Total',data=Yearwise)
```

```python
Yearwise.columns
```

```python

Yearwise.plot(x='Year',y=['POONDI', 'CHOLAVARAM', 'REDHILLS', 'CHEMBARAMBAKKAM'], figsize=(14,8))
```

```python
## Monthwise variation in the four dams
Monthwise = df.groupby('Month', as_index=False).mean()
Monthwise = Monthwise[['Month', 'POONDI', 'CHOLAVARAM', 'REDHILLS', 'CHEMBARAMBAKKAM','Total']]
Monthwise
```

```python
Monthwise.plot(x='Month',y=['POONDI', 'CHOLAVARAM', 'REDHILLS', 'CHEMBARAMBAKKAM'], figsize=(14,8))
```

```python
df.info()
```

```python
df['Year'] = df['Year'].astype('category')
df['Month'] = df['Month'].astype('category')
```

```python
df.info()
```

```python
## Water levels in the month of February for the all the years in total water level of the 4 dams

febwise = df[df['Month'] == 2]
febwise.head()
febwise = febwise.groupby('Year', as_index = False).mean()

febwise.plot(x='Year', y = ['POONDI', 'CHOLAVARAM', 'REDHILLS', 'CHEMBARAMBAKKAM'], figsize = (14, 8))
```

### Inferences:

* Although chennai receive some rainfall in July-Sep due to SW monsoon but majority of rainfall happens in Oct-Dec
* From 2006-10  NE monsoon has much more contribution than SW monsoon But trend reversed in last few years, either SW and NE are contributing equally or SW is contributing more than NE.
* Declining NE monsoon rainfall and thereby water level in four reservoir of Chennai in recent years.
