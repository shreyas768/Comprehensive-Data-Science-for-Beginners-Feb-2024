```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
def last_four(num):
    return int(str(num)[-4:])
```

```python
df['CC Number'][2]
```

```python
last_four(df['CC Number'][2])
```

```python
df['CC_last_four']=df['CC Number'].apply(last_four)
df.head()
```

```python
def yelp(price):
    if price < 10:
        return 'Low'
    elif price >9 and price <30:
        return 'medium'
    else:
        return 'Expensive'
```

```python
df['type']=df['total_bill'].apply(yelp)
```

```python
df.sample(10)
```

```python
def quality(bill,ti):
    if ti/bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
```

```python
df['tip_quality']=df[['total_bill','tip']].apply(lambda df : quality(df['total_bill'],df['tip']),axis=1)
```
